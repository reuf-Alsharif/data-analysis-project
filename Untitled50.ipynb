```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# تحميل البيانات
data = pd.read_csv('/content/diabetes.csv')

# عرض أول خمس صفوف للتعرف على هيكل البيانات
print(data.head())
```

```python
# عرض المعلومات العامة حول البيانات
print(data.info())

# عرض إحصاءات وصفية للبيانات
print(data.describe())
```

```python
# التحقق من القيم المفقودة
print(data.isnull().sum())

# إذا كانت هناك قيم مفقودة، يمكن ملؤها بمتوسط القيم كالتالي:
data.fillna(data.mean(), inplace=True)
```

```python
# التحقق من القيم الصفرية في الأعمدة
zero_columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for column in zero_columns:
    print(f"{column} zeros: {data[column].eq(0).sum()}")


# استبدال القيم الصفرية بمتوسط القيم في هذه الأعمدة
for column in zero_columns:
    data[column] = data[column].replace(0, data[column].mean())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# رسم Boxplot لكل عمود للتحقق من القيم الشاذة
plt.figure(figsize=(12, 8))
sns.boxplot(data=data)
plt.xticks(rotation=45)
plt.title("Boxplot for Outlier Detection")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# توحيد المقياس للبيانات
scaler = StandardScaler()
X = data.drop(columns=['Outcome'])
X_scaled = scaler.fit_transform(X)
y = data['Outcome']

# تحويل البيانات إلى DataFrame بعد التوحيد
X = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
# تعريف المتغيرات المستقلة (X) والمتغير التابع (y)
X = data.drop(columns=['Outcome'])
y = data['Outcome']
```

```python
# تقسيم البيانات بنسبة 80% للتدريب و20% للاختبار
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
# إنشاء نموذج Naive Bayes
nb_model = GaussianNB()

# تدريب النموذج باستخدام بيانات التدريب
nb_model.fit(X_train, y_train)
```

```python
# التنبؤ بالنتائج باستخدام بيانات الاختبار
y_pred = nb_model.predict(X_test)
```

```python
# حساب دقة النموذج
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# إنشاء مصفوفة الارتباك
conf_matrix = confusion_matrix(y_test, y_pred)

# رسم مصفوفة الارتباك
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=['No Diabetes', 'Diabetes'], yticklabels=['No Diabetes', 'Diabetes'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
# إنشاء تقرير التصنيف
class_report = classification_report(y_test, y_pred, target_names=['No Diabetes', 'Diabetes'])
print("Classification Report:")
print(class_report)
```

```python
# استيراد المكتبات الضرورية
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# تحميل البيانات
diabetes_data = pd.read_csv('/content/diabetes.csv')

# تجهيز المتغيرات X و y
X = diabetes_data.drop('Outcome', axis=1)  # حذف العمود المستهدف 'Outcome'
y = diabetes_data['Outcome']  # تحديد العمود المستهدف

# تقسيم البيانات
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# إنشاء وتدريب نموذج Gaussian Naive Bayes
gnb_model = GaussianNB()
gnb_model.fit(X_train, y_train)

# التنبؤ وحساب الدقة
y_pred_gnb = gnb_model.predict(X_test)
accuracy_gnb = accuracy_score(y_test, y_pred_gnb)
print(f"Gaussian Naive Bayes Model Accuracy: {accuracy_gnb:.2f}")
print(classification_report(y_test, y_pred_gnb, target_names=['No Diabetes', 'Diabetes']))

```

```python
from sklearn.naive_bayes import MultinomialNB

# إنشاء وتدريب نموذج Multinomial Naive Bayes
mnb_model = MultinomialNB()
mnb_model.fit(X_train, y_train)

# التنبؤ باستخدام بيانات الاختبار
y_pred_mnb = mnb_model.predict(X_test)

# تقرير التصنيف
print("Multinomial Naive Bayes Classification Report:")
print(classification_report(y_test, y_pred_mnb, target_names=['No Diabetes', 'Diabetes']))
```

```python
from sklearn.naive_bayes import BernoulliNB

# إنشاء وتدريب نموذج Bernoulli Naive Bayes
bnb_model = BernoulliNB()
bnb_model.fit(X_train, y_train)

# التنبؤ وحساب الدقة
y_pred_bnb = bnb_model.predict(X_test)
accuracy_bnb = accuracy_score(y_test, y_pred_bnb)
print(f"Bernoulli Naive Bayes Model Accuracy: {accuracy_bnb:.2f}")
print(classification_report(y_test, y_pred_bnb, target_names=['No Diabetes', 'Diabetes']))
```

```python
# استيراد المكتبات الضرورية
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# تحميل البيانات
diabetes_data = pd.read_csv('/content/diabetes.csv')

# تجهيز المتغيرات X و y
X = diabetes_data.drop('Outcome', axis=1)  # حذف العمود المستهدف 'Outcome'
y = diabetes_data['Outcome']  # تحديد العمود المستهدف

# تقسيم البيانات إلى مجموعة تدريب ومجموعة اختبار
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# إنشاء وتدريب نموذج Logistic Regression
logistic_model = LogisticRegression(max_iter=200, random_state=0)
logistic_model.fit(X_train, y_train)

# إجراء التنبؤات باستخدام بيانات الاختبار
y_pred_logistic = logistic_model.predict(X_test)

# حساب دقة النموذج
accuracy_logistic = accuracy_score(y_test, y_pred_logistic)
print(f"Logistic Regression Model Accuracy: {accuracy_logistic:.2f}")

# عرض تقرير التصنيف
print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_logistic, target_names=['No Diabetes', 'Diabetes']))

# رسم مصفوفة الارتباك
conf_matrix_logistic = confusion_matrix(y_test, y_pred_logistic)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_logistic, annot=True, fmt='d', cmap='Blues', xticklabels=['No Diabetes', 'Diabetes'], yticklabels=['No Diabetes', 'Diabetes'])
plt.title('Confusion Matrix for Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
